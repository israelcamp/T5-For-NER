```python
import json
import re
```

```python
def read_json(filepath):
    with open(filepath, 'rb') as f:
        return json.load(f)
```

```python
def clean_text(text):
    text = text.strip().replace('\n', ' ')
    text = re.sub(' {1,}', ' ', text)
    return text
```

```python
SELECTIVE_CATEGS = [
    'PESSOA',
    'ORGANIZACAO',
    'LOCAL',
    'TEMPO',
    'VALOR',
]

ENTITIES_MAP = {
    'PESSOA': 'PER',
    'ORGANIZACAO': 'ORG',
    'LOCAL': 'LOC',
    'TEMPO': 'TMP',
    'VALOR': 'VAL'
}
```

```python
harem_json = read_json('../data/harem/CDPrimeiroHAREMMiniHAREM-selective.json')
```

```python
examples = ['DOCSTART O']
for j, document in enumerate(harem_json):
    doc_text = document['doc_text']
    entities = document['entities']
    
    samples = []
    text_offset = 0
    for ent in entities:

        start_offset = ent['start_offset']
        end_offset = ent['end_offset']
        # samples before
        o_text = clean_text(doc_text[text_offset:start_offset])
        if len(o_text):
            o_text_split = o_text.split(' ')
            for t in o_text_split:
                samples.append(f'{t} O')

        # the entity
        label = ENTITIES_MAP[ent['label']]
        ent_text = clean_text(doc_text[start_offset:end_offset])
        ent_text_split = ent_text.split(' ')
        for i, t in enumerate(ent_text_split):
            tag = 'B' if i == 0  else 'I'
            samples.append(f'{t} {tag}-{label}')

        text_offset = end_offset
    
    # adding leftovers
    if text_offset < len(doc_text):
        o_text = clean_text(doc_text[text_offset:len(doc_text)])
        if len(o_text):
            o_text_split = o_text.split(' ')
            for t in o_text_split:
                samples.append(f'{t} O')
    
#     if j == 1:
#         break
    
    examples.append('\n'.join(samples))
```
